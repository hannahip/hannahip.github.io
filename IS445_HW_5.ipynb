```python
import altair as alt
import pandas as pd

# load data
data_url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
df = pd.read_csv(data_url)

alt.data_transformers.disable_max_rows()
```

```python
plot1 = alt.Chart(data_url).mark_bar().encode(
    x=alt.X('Agency Name:N', title='Agency', sort='-y'), # Sort by size
    y=alt.Y('sum(Square Footage):Q', title='Total Square Footage'),
    color=alt.value('darkgreen')
).properties(
    width=800,
    title="Total Square Footage Managed by State Agencies"
)

plot1
```

```python
# create dropdown
status_dropdown = alt.binding_select(options=df['Bldg Status'].unique().tolist(), name="Building Status ")
status_select = alt.selection_point(fields=['Bldg Status'], bind=status_dropdown)

plot2 = alt.Chart(data_url).mark_rect().encode(
    x=alt.X('Year Acquired:T', bin=alt.Bin(maxbins=40), title='Year Acquired'),
    y=alt.Y('count():Q', title='Number of Buildings'),
    color=alt.Color('Bldg Status:N', scale=alt.Scale(scheme='category10')),
    opacity=alt.condition(status_select, alt.value(1), alt.value(0.1))
).add_params(
    status_select
).properties(
    width=800,
    title="Building Acquisitions Over Time (Filter by Status)"
)

plot2
```

```python
plot1.save('plot1.json')
plot2.save('plot2.json')
```
